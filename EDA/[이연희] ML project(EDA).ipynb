## (3) 데이터 시각화

```python
# Target이 될 데이터 확인

sns.countplot(data=health, x='Heart_Disease')
plt.title('Counts of Heart Disease')
plt.show()
```

```python
len(health[health['Heart_Disease'] == 'Yes']), len(health[health['Heart_Disease'] == 'No'])
```

Yes/No 수치가 많이 차이남. 파이그래프로 비율 확인

```python
# 심장병 유무의 비율

hd_counts = health['Heart_Disease'].value_counts()

plt.pie(hd_counts, autopct='%1.1f%%',labels=hd_counts.index)
plt.axis('equal')
plt.title('Ratio of Heart Disease')
plt.show()
```

```python
# 다른 질병 유무들의 countplot

f, ax = plt.subplots(2,2, figsize=(15,10))

sns.countplot(data=health, x='Skin_Cancer', ax=ax[0,0], palette='Paired')
sns.countplot(data=health, x='Other_Cancer', ax=ax[0,1], palette='Paired')

sns.countplot(data=health, x='Depression', ax=ax[1,0], palette='Paired')
sns.countplot(data=health, x='Arthritis', ax=ax[1,1], palette='Paired')
plt.show()
```

```python
sc_counts = health['Skin_Cancer'].value_counts()
d_counts = health['Depression'].value_counts()
a_counts = health['Arthritis'].value_counts()
oc_counts = health['Other_Cancer'].value_counts()
```

```python
f, ax = plt.subplots(2, 2, figsize=(10,10))

ax[0,0].pie(sc_counts, autopct='%1.1f%%',labels=sc_counts.index)
ax[0,0].set_title('Ratio of Skin Cancer')

ax[0,1].pie(d_counts, autopct='%1.1f%%',labels=d_counts.index)
ax[0,1].set_title('Ratio of Depression')

ax[1,0].pie(a_counts, autopct='%1.1f%%',labels=a_counts.index)
ax[1,0].set_title('Ratio of Arthritis')

ax[1,1].pie(oc_counts, autopct='%1.1f%%',labels=oc_counts.index)
ax[1,1].set_title('Ratio of Other Cancer')

plt.show()
```

```python

diabetes_counts = health['Diabetes'].value_counts()

f, ax = plt.subplots(1,2, figsize=(18,8))

ax[0].pie(diabetes_counts, autopct='%1.1f%%', labels=diabetes_counts.index)
ax[0].set_title('Ratio of Diabetes')

sns.countplot(data=health, x='Diabetes', palette='pastel', ax=ax[1])
ax[1].set_title('Count of Diabetes')
ax[0].set_ylabel('')

plt.show()
```

#### 'Yes, but female told only during pregnancy', 'No, pre-diabetes or borderline diabetes' 비율이 너무 낮아 의미 없다고 판단하여 각각 Yes/No에 포함하기로 판단함

```python
health['Diabetes'] = health['Diabetes'].replace('Yes, but female told only during pregnancy', 'Yes')
health['Diabetes'] = health['Diabetes'].replace('No, pre-diabetes or borderline diabetes', 'No')
```

```python
health['Diabetes'].unique()
```

```python
diabetes_counts = health['Diabetes'].value_counts()

f, ax = plt.subplots(1,2, figsize=(18,8))

ax[0].pie(diabetes_counts, autopct='%1.1f%%', labels=diabetes_counts.index)
ax[0].set_title('Ratio of Diabetes')

sns.countplot(data=health, x='Diabetes', ax=ax[1])
ax[1].set_title('Counts of Diabetes')
ax[0].set_ylabel('')

plt.show()
```
